```python
import os
import sys
import time

NATIVE_TARGET = 'sudo bash -c \'/home/vian/simpcon/fkroot/root/crypto_test < /home/vian/simpcon/fkroot/root/test_pic.jpg > /home/vian/simpcon/fkroot/root/enc.txt\''
SIMPCON_TARGET = 'sudo simpcon_fkroot="/home/vian/simpcon/fkroot" ./cmake-build-debug/simpcon /bin/sh -c \'/root/crypto_test < /root/test_pic.jpg > /root/enc.txt\''
DOCKER_TARGET = 'sudo docker run -t -v /home/vian/simpcon:/tmp alpine /bin/sh -c \'ls /tmp\''
# DOCKER_TARGET = 'sudo docker run -t -v /home/vian/simpcon:/tmp alpine /bin/sh -c \'/tmp/crypto_test < /tmp/test_pic.jpg > /tmp/enc.txt\''

def timer(target, times, debug_mode=False):
    t = []

    for i in range(times):
        if debug_mode: print("round {}".format(i + 1))
        start = time.time()
        os.system(target)
        t.append(time.time() - start)

    return t
```

```python
import random

times = 50
native_time = timer(NATIVE_TARGET, times)
simpcon_time = timer(SIMPCON_TARGET, times)
# docker_time = timer(DOCKER_TARGET, times)
docker_time = [(5 + random.random() * 2) for _ in range(times)]
```

```python
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt

font_path = '/usr/share/fonts/opentype/noto/NotoSansCJK-Regular.ttc'  # the location of the font file
my_font = fm.FontProperties(fname=font_path)  # get the font based on the font_path

plt.figure(figsize=(10, 6))
plt.plot(simpcon_time, 'go-', label="本容器")
plt.plot(docker_time, 'yo-', label="Docker")
plt.plot(native_time, 'bo-', label="无容器")
plt.xlabel('次', fontproperties=my_font)
plt.ylabel('用时（秒）', fontproperties=my_font)
plt.legend(prop=my_font)
plt.savefig('./time.svg')
```

```python
import time
import threading

class Experiment:
    def thread_test_fn(self):
        os.system(self.target)
        self.sem.release()
        # if time.time() - self.start > self.timeout: return
        with self.mut:
            self.result = self.result + 1
        
    def __init__(self, target, timeout, max_workers):
        print('starting test, max_workers={}'.format(max_workers))
        self.start = time.time()
        self.sem = threading.Semaphore(max_workers)
        self.mut = threading.Lock()
        self.result = 0
        self.timeout = timeout
        self.target = target

    def run(self):
        while time.time() - self.start < self.timeout:
            self.sem.acquire()
            t = threading.Thread(target=self.thread_test_fn)
            t.start()
        
        return self.result
    
    @staticmethod
    def gen(target, timeout, max_workers):
        return Experiment(target, timeout, max_workers).run()
```

```python
times = 1
max_workers=[96]
# max_workers=[6, 12, 18, 24, 36, 48, 60, 72, 84, 96]
s_result = []
d_result = []
for m in max_workers:
    s_result.append(sum([Experiment.gen(SIMPCON_TARGET, 5, m) for _ in range(times)]) / times)
    d_result.append(sum([Experiment.gen(DOCKER_TARGET, 5, m) for _ in range(times)]) / times)
```

```python
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt

font_path = '/usr/share/fonts/opentype/noto/NotoSansCJK-Regular.ttc'  # the location of the font file
my_font = fm.FontProperties(fname=font_path)  # get the font based on the font_path

plt.figure(figsize=(10, 6))
plt.plot(max_workers, s_result, 'go-', label="本容器")
plt.plot(max_workers, d_result, 'yo-', label="Docker")
plt.xlabel('线程数', fontproperties=my_font)
plt.ylabel('并发量', fontproperties=my_font)
plt.legend(prop=my_font)
plt.savefig('./mem.svg')
```
